# Meandering river

```python
%matplotlib inline
import numpy as np
import ttim
```

```python
ml = ttim.Model3D(
    kaq=[2, 1, 5, 10, 4],
    z=[10, 8, 6, 4, 2, 0],
    Saq=[0.1, 0.0001, 0.0002, 0.0002, 0.0001],
    phreatictop=True,
    kzoverkh=0.1,
    topboundary="conf",
    tmin=1,
    tmax=10,
)
w = ttim.Well(ml, -25, 0, rw=0.3, tsandQ=[(0, 100)], layers=[2, 3])

dxdy = 30 * np.pi / 100 * np.cos(np.pi)
x1 = np.arange(-150, -100, 10)
y1 = (x1 + 100) * dxdy
x2 = np.linspace(-100, 100, 20)
y2 = 30 * np.sin(x2 * np.pi / 100)
yls = np.hstack((y1, y2))
xls = np.hstack((x1, x2))
xy = np.array([xls, yls]).T

ls1 = ttim.HeadLineSinkString(ml, xy=xy, tsandh="fixed", layers=[0, 1])
```

```python
ml.solve()
```
