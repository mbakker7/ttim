### Modeling a higher order head line-sink with variable head

```python
import matplotlib.pyplot as plt
import numpy as np

import ttim

plt.rcParams["font.size"] = 8.0
plt.rcParams["figure.figsize"] = (4, 3)
```

```python
ml = ttim.ModelMaq(kaq=10, z=[11, 10, 0], Saq=1e-4, c=1000, topboundary='semi', tmin=0.01, tmax=10)
hls = ttim.HeadLineSinkHo(model=ml, x1=-1000, y1=-500, x2=1000, y2=500, 
                          tsandh=[(0, 2), (5, 4)], order=5)
ml.solve()
```

```python
t = np.linspace(0.01, 10, 100)
h = ml.head(x=-200, y=400, t=t)
plt.plot(t, h[0])
plt.grid()
```
