### Test d1d2 function

```python
import matplotlib.pyplot as plt
import numpy as np

import ttim
```

```python
ml1 = ttim.ModelMaq(
    kaq=10, z=[11, 10, 0], c=100, Saq=0.001, topboundary="semi", tmin=1, tmax=10
)
ls1 = ttim.HeadLineSink(model=ml1, x1=-100, y1=0, x2=100, y2=0, tsandh=[(0, 1)])
ml1.solve()
```

```python
ml2 = ttim.ModelMaq(
    kaq=10, z=[11, 10, 0], c=100, Saq=0.001, topboundary="semi", tmin=1, tmax=10
)
ls2 = ttim.HeadLineSinkHo(
    model=ml2, x1=-100, y1=0, x2=100, y2=0, tsandh=[(0, 1)], order=0
)
ml2.solve()
```

```python
ylim = 30 * ml1.aq.lababs.item()
N = 100
x = np.zeros(N)
y = np.linspace(0, ylim, N)
h1 = ml1.headalongline(x, y, 10).squeeze()
h2 = ml2.headalongline(x, y, 10).squeeze()
plt.semilogx(y[1:], h1[1:])
plt.semilogx(y[1:], h1[1:], "--")
```
